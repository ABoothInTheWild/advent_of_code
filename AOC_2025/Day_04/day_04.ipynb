## Part 1

```python
# Test data
test = """..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@."""

test_grid = []
for line in test.split("\n"):
    test_grid.append(list(line))

test_rows = len(test_grid)
test_columns = len(test_grid[0])

test_rolls = 0
for i in range(test_rows):
    for j in range(test_columns):
        if test_grid[i][j] == "@":
            sum_of_adjacent = 0
            # Check all 8 adjacent positions
            for di in [-1, 0, 1]:
                for dj in [-1, 0, 1]:
                    if di == 0 and dj == 0:
                        continue  # Skip the current cell
                    ni, nj = i + di, j + dj
                    if 0 <= ni < test_rows and 0 <= nj < test_columns:
                        if test_grid[ni][nj] == "@":
                            sum_of_adjacent += 1

            if sum_of_adjacent < 4:
                test_rolls += 1

print(f"Test result: {test_rolls}")
```

## Part 2

```python
def count_rolls(grid):
    rows = len(grid)
    columns = len(grid[0])
    total_rolls = 0
    vals_to_replace = set()

    for i in range(rows):
        for j in range(columns):
            if grid[i][j] == "@":
                sum_of_adjacent = 0
                # Check all 8 adjacent positions
                for di in [-1, 0, 1]:
                    for dj in [-1, 0, 1]:
                        if di == 0 and dj == 0:
                            continue  # Skip the current cell
                        ni, nj = i + di, j + dj
                        if 0 <= ni < rows and 0 <= nj < columns:
                            if grid[ni][nj] == "@":
                                sum_of_adjacent += 1

                if sum_of_adjacent < 4:
                    vals_to_replace.add((i, j))
                    total_rolls += 1

    new_grid = [row[:] for row in grid]
    for i, j in vals_to_replace:
        new_grid[i][j] = "."

    return total_rolls, new_grid
```

```python
# Test data
test = """..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@."""

test_grid = []
for line in test.split("\n"):
    test_grid.append(list(line))

test_total_rolls = 0
test_iter_rolls = -1
test_new_grid = test_grid

while test_iter_rolls != 0:
    test_iter_rolls, test_new_grid = count_rolls(test_new_grid)
    test_total_rolls += test_iter_rolls

print(f"Test result: {test_total_rolls}")
```

```python
with open("input2.txt", "r") as file:
    data = file.read().strip()

grid = []
for line in data.split("\n"):
    grid.append(list(line))

total_rolls = 0
iter_rolls = -1
new_grid = grid

while iter_rolls != 0:
    iter_rolls, new_grid = count_rolls(new_grid)
    total_rolls += iter_rolls

print(f"Part 2 answer: {total_rolls}")
```
