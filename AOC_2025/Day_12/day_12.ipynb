# Advent of Code 2025 - Day 12

```python
# Test data
test = """4x4: 0 0 0 0 2 0
12x5: 1 0 1 0 2 2
12x5: 1 0 1 0 3 2"""

test_lines = [line.strip().split() for line in test.splitlines() if "x" in line]

test_data = []
for parts in test_lines:
    size = [int(x) for x in parts[0][:-1].split("x")]
    area = size[0] * size[1]
    presents = sum([int(x) for x in parts[1:]])
    test_data.append((size, area, presents))

print(f"Test rows: {len(test_data)}")
```

## Part 1

```python
def count_valid_boxes(data):
    """
    Count boxes where area >= presents * 9.
    """
    count = 0
    for size, area, presents in data:
        if area >= presents * 9:
            count += 1
    return count
```

```python
# Test Part 1
result = count_valid_boxes(test_data)
print(f"Test result: {result}")
```

```python
# Load real input
with open("input.txt", "r") as f:
    lines = [line.strip().split() for line in f if "x" in line]

real_data = []
for parts in lines:
    size = [int(x) for x in parts[0][:-1].split("x")]
    area = size[0] * size[1]
    presents = sum([int(x) for x in parts[1:]])
    real_data.append((size, area, presents))

print(f"Input rows: {len(real_data)}")
```

```python
# Part 1 - Real input
result = count_valid_boxes(real_data)
print(f"Part 1 result: {result}")
result
```
