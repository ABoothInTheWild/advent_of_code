```python
import numpy as np
import pandas as pd

# Test data
test = """123 328  51 64
 45 64  387 23
  6 98  215 314
*   +   *   +  """
```

```python
## Part 1
```

```python
# Test example
test_rows = test.split("\n")
test_rows = [row.strip().split() for row in test_rows]
df = pd.DataFrame(test_rows)

final_result = 0
for col in df.columns:
    vals = df.loc[0:len(df), col].values
    nums = vals[0:len(vals)-1]
    nums = [int(n) for n in nums]
    operator = vals[-1]

    if operator == "+":
        result = np.sum(nums)
    elif operator == "*":
        result = np.prod(nums)

    final_result += result

print(f"Test result: {final_result}")
```

```python
# Test example
test_rows = test.split("\n")
width = max(len(r) for r in test_rows)
grid = [list(r.ljust(width)) for r in test_rows]
grid = np.array(grid)
```

```python
blank = [all(grid[r][c] == " " for r in range(len(grid)))
         for c in range(grid.shape[1])]

segments = []
current = []

for c, b in enumerate(blank):
    if b:
        if current:
            segments.append(current)
            current = []
    else:
        current.append(c)

if current:
    segments.append(current)

# Right-to-left problem order
segments = segments[::-1]

cols = []

for seg in segments:
    # Collect each number: read column top→bottom, ignore spaces
    nums = []
    for c in seg:
        digits = [grid[r][c] for r in range(len(grid)-1) if grid[r][c] != " "]
        nums.append("".join(digits))

    # operator (bottom row)
    op = None
    for c in seg:
        if grid[-1][c] != " ":
            op = grid[-1][c]
            break

    # Reverse nums because numbers are read right→left within the problem
    nums = nums[::-1]

    # Convert numeric strings to int
    nums = [int(n) for n in nums]

    # Build a single column: numbers followed by operator
    col = nums + [op]
    cols.append(col)

# Normalize column lengths
max_len = max(len(c) for c in cols)
for i in range(len(cols)):
    while len(cols[i]) < max_len:
        cols[i].insert(0, None)  # top padding if needed

final_result = 0

for col in cols:
    # col is a list: [None, None, ..., num1, num2, ..., operator]
    # Filter out None values and separate numbers from operator
    vals = [v for v in col if v is not None]

    nums = vals[:-1]  # All but last are numbers
    nums = [int(n) for n in nums]

    operator = vals[-1]  # Last is the operator

    if operator == "+":
        result = np.sum(nums)
    elif operator == "*":
        result = np.prod(nums)

    final_result += result

print(f"Test result: {final_result}")
```
